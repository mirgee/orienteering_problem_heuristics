```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,5))

perfect_solved_x = [1997, 2007]
perfect_solved_y = [13, 19]
perfect_solved = ["Connect Four", "Chinook"]

imperfect_solved_x = [2005, 2015]
imperfect_solved_y = [12, 18]
imperfect_solved = ["Rhode Island Poker", "Cepheus"]

perfect_beats_human_x = [1996, 2016]
perfect_beats_human_y = [120, 170]
perfect_beats_human = ["Deep Blue", "AlphaGo"]

imperfect_beats_human_x = [2017]
imperfect_beats_human_y = [161]
imperfect_beats_human = ["DeepStack"]

ax = plt.gca()
ax.set_xlim([1995,2020])
ax.set_ylim([0,190])
ax.set_xlabel('year')
ax.set_ylabel('game tree size')

ax.scatter(perfect_solved_x, perfect_solved_y, color="blue", 
           label="perf. inf. solved", s=120, alpha=0.5)
for i, txt in enumerate(perfect_solved):
    ax.annotate(txt, (perfect_solved_x[i],perfect_solved_y[i]))

ax.scatter(imperfect_solved_x, imperfect_solved_y, color="darkred", 
           label="imperf. inf. solved", s=120, alpha=0.5)
for i, txt in enumerate(imperfect_solved):
    ax.annotate(txt, (imperfect_solved_x[i],imperfect_solved_y[i]))
    
ax.scatter(perfect_beats_human_x, perfect_beats_human_y, color="lightblue", 
           label="perf. inf. beats human", s=120)
for i, txt in enumerate(perfect_beats_human):
    ax.annotate(txt, (perfect_beats_human_x[i],perfect_beats_human_y[i]))
    
ax.scatter(imperfect_beats_human_x, imperfect_beats_human_y, color="red", 
           label="imperf. inf. beats human", s=120)
for i, txt in enumerate(imperfect_beats_human):
    ax.annotate(txt, (imperfect_beats_human_x[i],imperfect_beats_human_y[i]))
    
plt.legend(numpoints=1, loc="upper left")
plt.show()
fig.savefig('games.png')
```
